```python
import json
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
# url of the web page
scrape_url= "https://understat.com/league/EPL"
```

```python
urlopen(scrape_url)
```

```python
page_connect=urlopen(scrape_url)
page_connect
```

```python
page_html=BeautifulSoup(page_connect,"html.parser")
```

```python
page_html
```

```python
page_html.findAll(name="script")[3].text[32:-3]
```

```python
json_raw_str=page_html.findAll(name="script")[3].text
```

```python
json_raw_str.index("')")
```

```python
json_raw_str[32:297716]
```

```python
json_data=json_raw_str[32:297716]
```

```python
json_data=json_data.encode("utf8").decode("unicode_escape")
```

```python
json.loads(json_data)
```
